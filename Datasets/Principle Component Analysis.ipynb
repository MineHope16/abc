# PRINCIPLE COMPONENT ANALYSIS (PCA)

* Only work on numerical values
* Dataset must be preprocessed(no null values, same degree of understanfing)
* Standard Scaler = (x - X)/S.D  here, x means values of the features and X means Mean of the features



Note: No similar attrs should select

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#import the Datasets
data = pd.read_csv("driver-data.csv")
print(data.head())
print(data.info())
```

```python
#Feature Scaling
x = data.values
print(x)
x = StandardScaler().fit_transform(x)
print(x)
```

```python
#Creating an instance of the PCA class with component=2
pca = PCA(n_components=2)
principle_component = pca.fit_transform(x)
```

```python
#Creating a new dataset for PCA
principle_data = pd.DataFrame(data=principle_component, columns=['(Principle Component 1)','(Principle Component 2)'])
print(principle_data)
```

```python
plt.figure(figsize=(8,9))
plt.scatter(principle_data['(Principle Component 1)'], principle_data['(Principle Component 2)'])
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.title('PCA Visualization')
plt.show()
```

```python
# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
print("Total Explained Variance:", np.sum(explained_variance_ratio))
```
