# MATPLOTLIB - Python Visualization

```python
import matplotlib.pyplot as plp
import pandas as pd

cars = pd.read_csv("C:/Users/USER/Downloads/mtcars.csv")
```

```python
x = [2,4,5]
y = [3,5,7]
plp.plot(x,y)
plp.title("My Graph")
plp.legend(['Line'])
plp.show()
```

```python
plp.bar(x,y)
plp.title("My Bar Graph")
plp.legend(["Bar"])
plp.show()
```

```python
x = [2,3,4,5,2,3,6,7,8,4,5,6,0]

plp.hist(x, bins = [1,2,3,4,5,6,7])
plp.legend(['Bar'])
plp.title("My Histogram")
```

```python
x = [3,4,2,3,2,5,3,7,6,8]
y = [4,5,6,7,3,5,2,4,5,8]

plp.scatter(x,y)
plp.title("My Scatter Plot")
plp.legend(["Plots"])
plp.show()
```
